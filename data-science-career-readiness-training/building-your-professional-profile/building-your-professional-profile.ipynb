Building your professional profile and maintaining your online presence can be beneficial while you're looking for jobs. It's also going to help you later in your career, allowing you to be found by employers and recruiters. In this tutorial, we'll look at the following platforms:

* LinkedIn
* GitHub
* Medium
* Jovian

## LinkedIn

[Sign up](https://linkedin.com) | [Profile Guidelines](https://business.linkedin.com/en-uk/marketing-solutions/blog/posts/content-marketing/2017/17-steps-to-a-better-LinkedIn-profile-in-2017)

<img src="https://i.imgur.com/Am0gCBR.png" width="480">



### Why it Matters

* It's a place to record your education & professional history
* You will link to it from your Resume (it's like an extended Resume)
* Potential employers will look at your LinkedIn profile
* When you reach out for referrals, people will review your profile
* You will apply for jobs on LinkedIn & via LinkedIn on other sites
* Recruiters can find your profile by searching for skills & job roles

### What to Do

* Create a LinkedIn profile (if not created)
* Fill out your LinkedIn profile (see below)
    * Add a 2nd language, if useful
* Import your email contacts (connect but don’t invite)
* Set up job alerts for your target roles
* Share your profile in the Slack channel
* Connect with everyone in the bootcamp
* Endorse each other for data science skills
* Make sure your profile is publicly visible 
    * Make sure people can connect with you (“Who can reach you”)
* Add a Resume in “Job Application Settings”
* (Optional) Share your projects, blog posts and other work-related updates
* (Optional) Join your university groups

## GitHub

Sign up here: https://github.com

<img src="https://i.imgur.com/7DXUiof.png" width="640">



### Why it Matters

* Its a single place to show case all your programming work
* You will link to your GitHub profile from your Resume
* Recruiter & employers will often check your GitHub profile
* Your contributions to open-source projects show up here
* The README can be used to replicate your Resume


### What to Do

* Create a GitHub profile (if not created)
* Fill out your GitHub profile
* Create a [profile README](https://docs.github.com/en/github/setting-up-and-managing-your-github-profile/customizing-your-profile/managing-your-profile-readme)
* Create a repository for each of your projects
    * Upload the Jupyter notebook
    * Add a README (with Jovian & notebook link)
* Pin your best projects to your profile
* Share your profile on the channel
* Follow each other on GitHub
* Star each others’ projects
* (Optional) Contribute to an open source project

## Medium

Sign up here: https://medium.com

<img src="https://i.imgur.com/Q10VJio.png" width="480">



### Why it Matters

* It's a single place to publish all your written work (projects, tutorials, articles)
* Potential employers are more likely to read blog posts (compared to code repos)
* Medium posts are indexed by Google & show up in search results
* Medium posts can be submitted to publications & discovered on Medium
* It has a great writing interface, and can help you cultivate a writing habit
* Your Medium page can also become your personal website


### What to Do

* Create a Medium profile
* Use your full name
* Add a profile picture
* Add a Bio
* Select username
* (Optional) Set up a custom domain
* (Optional) Modify the layout
* (Optional) Enable email subscription
* Write a blog post for each project
    * Title & subtitle
    * Cover image
    * Introduction & outline
    * Sections (explanation & code)
    * Summary, References & Future Work
    * Tags
* Improve writing using Grammarly
* Don’t put your stories behind a paywall
* Submit your blogs to publications
    * Towards Data Science - https://towardsdatascience.com
    * Jovian - https://blog.jovian.ai
* Try to publish at least 3 blog posts (monthly once is good)
