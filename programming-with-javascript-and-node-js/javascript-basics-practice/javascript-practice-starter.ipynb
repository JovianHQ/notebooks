````javascript
sadWords = ['sad', 'bad', 'tragic', 'unhappy', 'worst']
````

To identify whether a tweet is happy, we can simply check if contains any of the words from `happy_words`. Here's an example:

````javascript
sampleTweet = tweets[0]
````

````javascript
sampleTweet
````

````javascript
isTweetHappy = false

// Get a word from happy_words
happyWords.forEach(word => {
    // Check if the tweet contains the word
    if (sampleTweet.includes(word)) {
        // Word found! Mark the tweet as happy
        isTweetHappy = true
    }
})
````

Do you understand what we're doing above? 

> For each word in the array of happy words, we check if is a part of the selected tweet. If the word is indeed a part of the tweet, we set the variable `isTweetHappy` to `true`.
